```python
df=df.drop( [' Date',' Frequency','longitude', 'latitude'],axis=1)
df.head()
```

```python
df.columns=['State', ' Estimated Unemployment Rate (%)',
       ' Estimated Employed', ' Estimated Labour Participation Rate (%)',
       'Region']
df.head()
```

```python
x=df['State']
```

```python
y=df[' Estimated Unemployment Rate (%)']
```

```python
y
```

```python
plt.bar(df['Region'],y,color='purple')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate')
plt.title('Estimated Unemployment Rate Region Wise')
plt.show()
```

```python
plt.scatter(df['Region'],y,color='Blue')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate')
plt.title('Estimated Unemployment Rate Region Wise by scatter plot')
plt.show()
```

```python
fig=px.bar(df, x='State', y=' Estimated Unemployment Rate (%)', color='State', title='Estimated Unemployment Rate State Wise')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
fig = px.sunburst(df, path=['Region', 'State'], values=' Estimated Unemployment Rate (%)')
fig.show()
```
